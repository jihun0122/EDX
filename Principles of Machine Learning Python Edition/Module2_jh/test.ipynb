```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credit = pd.read_csv('German_Credit.csv', header=None)
credit.columns = ['customer_id',
                  'checking_account_status', 'loan_duration_mo', 'credit_history', 
                  'purpose', 'loan_amount', 'savings_account_balance', 
                  'time_employed_yrs', 'payment_pcnt_income','gender_status', 
                  'other_signators', 'time_in_residence', 'property', 'age_yrs',
                  'other_credit_outstanding', 'home_ownership', 'number_loans', 
                  'job_category', 'dependents', 'telephone', 'foreign_worker', 
                  'bad_credit']
print(credit.shape)
credit.head()
```

```python
num_cols = ['loan_duration_mo', 'loan_amount', 'payment_pcnt_income',
            'age_yrs', 'number_loans', 'dependents']
```

```python
code_list=[['bad_credit',{2:1,1:0}]]
```

```python
for col_dic in code_list:
    col=col_dic[0]
    dic=col_dic[1]
    credit[col]=[dic[x] for x in credit[col]]

credit.head()
```

```python
def box_plot(credit,col_name,col_x='bad_credit'):
    for col in col_name:
        sns.set_style("whitegrid")
        sns.boxplot(col_x,col,data=credit)
        plt.xlabel(col_x)
        plt.ylabel(col)
        plt.show()

box_plot(credit,num_cols)
```

```python
auto_prices = pd.read_csv('Automobile price data _Raw_.csv')
```

```python
auto_prices.head()
```

```python
auto_prices.columns
```

```python
cols = ['price', 'bore', 'stroke', 
          'horsepower', 'peak_rpm']
```

```python
for columns in cols:
    auto_prices.loc[auto_prices[columns]=='?',columns]=np.nan
auto_prices.dropna(axis=0,inplace=True)
    
```

```python
auto_prices.head()
```

```python
for columns in cols:
    auto_prices[columns]=pd.to_numeric(auto_prices[columns])
```

```python
auto_prices.dtypes
```

```python
auto_prices.describe()
```

```python
cat_cols = ['make', 'fuel_type', 'aspiration', 'num_of_doors', 'body_style', 
            'drive_wheels', 'engine_location', 'engine_type', 'num_of_cylinders', 
            'fuel_system']
```

```python
for col in cat_cols:
    print('\n'+col)
    print(auto_prices[col].value_counts())
```

```python
for col in cat_cols:
    counts=auto_prices[col].value_counts()
    counts.plot.bar()
    plt.title(col)
    plt.show()
```

```python
num_cols = [ 'city_mpg', 'price']   
for col in num_cols:
    auto_prices[col].plot.hist(bins=10)
    plt.title(col+'hist')
    plt.show()
```

```python
for col in num_cols:
    sns.set_style('whitegrid')
    sns.distplot(auto_prices[col],bins=10,hist=True)
    plt.show()
```

```python
for col in num_cols:
    auto_prices.plot.scatter(x=col,y='price')
    plt.show()
```
